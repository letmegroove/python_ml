PANDAS 는 2차원 데이터 핸들링에 가장 많이 씀

특히, 시계열 데이터 핸들링에 많은 기능을 제공함

- Dataframe: column X Row 2차원 데이터 셋 (ndarray 로 맹글어짐)

- Series: 1개의 Column 값으로만 구성된 1차원 데이터셋

```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('titanic_train.csv')
print('titanic 변수 type: ', type(titanic_df))
```

```python
#head...DataFrame 의 맨 앞 일부 데이터만 추출
#맨 앞 0 1 2 3 4 는 일종의 PK / RowNum. 컬럼명은 따로 없음
titanic_df.head()
```

```python
#DataFrame 생성

dic1 = {'Name': ['Travis', 'Lucas', 'Neo', 'Steven'],
        'Year': [2011, 2013, 2009, 2019],
        'Gender': ['Male','Male','Female','Male']}

#Dictionary 를 DataFrame으로 변환
data_df = pd.DataFrame(dic1)
print(data_df)
print('#'*30)

#새로운 컬럼명 추가
data_df = pd.DataFrame(dic1, columns=['Name','Year','Gender','Age'])
print(data_df)
print('#'*30)

#Index를 새로운 값으로
data_df = pd.DataFrame(dic1, index=['one','two','three','four'])
print(data_df)
print('#'*30)
```

```python
print('columns: \n', titanic_df.columns)
print('columns: \n', titanic_df.index)
print('columns: \n', titanic_df.index.values)
```

```python
#Series

#1차원
series = titanic_df['Name']
print(series.head())
print(type(series))

#2차원
series2 = titanic_df[['Name','Age']]
print(series2.head())
print(type(series2))

#컬럼이 하나인 2차원
series3 = titanic_df[['Name']]
print(series3.head())
print(type(series3))
```

```python
#shape
#index는 컬럼에 포함되지 않음

print('Shape:', titanic_df.shape)
```

```python
#info...메타 데이터 정보를 보여줌
#memory usage를 잘 확인하는 버릇 만들것

titanic_df.info()
```

```python
#describe... 평균, 표준편차, mix/max 등의 정보를 보여줌

titanic_df.describe()
```

```python
#value_counts()...데이터 값의 분포도 확인 시 많이 쓰임....Series 객체에서만 호출될 수 있음을 유의

value_count = titanic_df['Pclass'].value_counts()
print(value_count)
print(type(value_count))
```

```python
#sort_values()...by 에 정렬컬럼 ascending 의 디폴트는 오름차순.. 내림차순 원할 시 False

titanic_df.sort_values(by='Age', ascending=False)
```

```python
#선별된 Column으로만 보여주는 것도 가능
titanic_df[['Name','Age']].sort_values(by='Age', ascending=False)
```

```python
#다수 컬럼으로 정렬도 가능
titanic_df[['Name','Age','Pclass']].sort_values(by=['Age','Pclass'], ascending=False)
```
